```python
import logging
logging.basicConfig(filename = "Test.log", level = logging.INFO, format = '%(asctime)s %(levelname)s %(message)s')
def sum_list(l):
  logging.info("Initiated sum_list function")
  try:
    print(f"Sum of numbers present in the list: {sum(l)}")
  except Exception as e:
    logging.error("An error has occurred")
    logging.exception("Exception is "+ str(e))

def prod_list(l):
  prod = 1
  logging.info("Required variables initialized")
  try:
    logging.info("Performing multiplication operation on the given list")
    for i in l:
      prod *= i
    logging.info("Logging product of the list"+ str(prod))
    print(f"Product of numbers in the list {prod}")
  except Exception as e:
    logging.error("An error has occurred")
    logging.exception("Exception is "+ str(e))

def smallest(l):
  logging.info("Executing minimum function")
  try:
    logging.info("Logging the smallest number after "+ str(min(l)))
    print(f"Smallest number of the list is {min(l)}")
  except Exception as e:
    logging.error("An error has occurred")
    logging.exception("Exception is "+ str(e))

def largest(l):
  logging.info("Executing largest function")
  try:
    logging.info("Logging largest number of the list "+ str(max(l)))
    print(f"Largest number of the list is {max(l)}")
  except Exception as e:
    logging.error("An error has occurred")
    logging.exception("Exception is "+ str(e))

def second_largest(l):
  logging.info("Executing second largest function")
  try:
    logging.info("Cloning the list")
    # Cloning operation represented here
    l1 = cloning(l)
    logging.info("Removing the maximum number")
    l1.remove(max(l1))
    logging.info("Logging second largest number "+ str(max(l1)))
    print(f"Second largest number of the list {max(l1)}")
  except Exception as e:
    logging.error("An error has occurred")
    logging.exception("Exception is "+ str(e))

def tot_large_num(l):
  logging.info("Executing tot_large_num function")
  try:
    print(f"{max(l)} is the largest with count {l.count(max(l))}")
    logging.info("Execution successful")
  except Exception as e:
    logging.error("An error has occurred")
    logging.exception("Exception is "+ str(e))

def even(l):
  logging.info("Executing even function")
  try:
    logging.info("Printing even numbers present in the list")
    print("Even numbers of the list")
    for i in l:
      if i%2 == 0:
        print(f"{i}")
    logging.info("Execution successful")
  except Exception as e:
    logging.error("An error has occurred")
    logging.exception("Exception is "+ str(e))

def odd(l):
  logging.info("Executing odd function")
  try:
    logging.info("Printing odd numbers present in the list")
    print("Odd numbers of the list")
    for i in l:
      if i%2 != 0:
        print(f"{i}")
    logging.info("Execution successful")
  except Exception as e:
    logging.error("An error has occurred")
    logging.exception("Exception is "+ str(e))

def rem_emp_list(l):
  logging.info("Executing rem_emp_list function")
  try:
    logging.info("Removing empty list from the list and printing the result")
    print(f"Before removing []\nf{l}")
    l.remove([])
    print("After removing []")
    print(l)
    logging.info("Execution successful")
  except Exception as e:
    logging.error("An error has occurred")
    logging.exception("Exception is "+ str(e))

def cloning(l):
  logging.info("Executing cloning function")
  try:
    logging.info("Cloning started")
    return l.copy()
    logging.info("Cloning successful")
  except Exception as e:
    logging.error("An error has occurred")
    logging.exception("Exception is "+ str(e))

def list_count(l):
  logging.info("Executing list_count function")
  try:
    logging.info("Counting operation commenced")
    for i in l:
     print(f"Count of number {i} is {l.count(i)}")
     logging.info("Execution successful")
  except Exception as e:
    logging.error("An error has occurred")
    logging.exception("Exception is "+ str(e))
```
